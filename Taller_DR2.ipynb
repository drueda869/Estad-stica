```python
# Punto 1 - Intervalo de confianza del 95% para la media (σ conocida)
# Enunciado: n=50, x̄=170 cm, σ=10 cm. Construir IC 95% para la media poblacional.

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# --------------------------
# 1) Datos del problema
n = 50                   # tamaño de la muestra
x_bar = 170.0            # media muestral (cm)
sigma = 10.0             # desviación estándar poblacional (σ conocida)
confianza = 0.95         # nivel de confianza

# --------------------------
# 2) Parámetros y fórmula
alpha = 1 - confianza
se = sigma / np.sqrt(n)  # error estándar de la media

# Para σ conocida usamos la distribución normal Z
z_crit = stats.norm.ppf(1 - alpha/2)   # valor crítico Z para intervalo bilateral

# Margen de error e intervalo de confianza
margen = z_crit * se
ci_inferior = x_bar - margen
ci_superior = x_bar + margen

# --------------------------
# 3) Imprimir resultados paso a paso
print("PUNTO 1 - INTERVALO DE CONFIANZA (σ conocida)")
print("--------------------------------------------------")
print(f"Datos: n = {n}, x̄ = {x_bar:.2f} cm, σ = {sigma:.2f} cm")
print(f"Nivel de confianza = {confianza*100:.0f}%  →  α = {alpha:.3f}")
print(f"Error estándar (SE) = σ/√n = {se:.4f} cm")
print(f"Valor crítico Z (bilateral) = {z_crit:.4f}")
print(f"Margen de error = Z * SE = {margen:.4f} cm")
print(f"Intervalo de confianza del {confianza*100:.0f}%: [{ci_inferior:.4f} cm, {ci_superior:.4f} cm]")
print("Interpretación: Con un 95% de confianza, la media poblacional de la altura se encuentra dentro de este intervalo.")

# --------------------------
# 4) Gráfica de la distribución muestral de la media (normal) y el IC
# Distribución de x̄ ~ N(μ ≈ x̄, SE)
x = np.linspace(x_bar - 4*se, x_bar + 4*se, 500)
pdf = stats.norm.pdf(x, loc=x_bar, scale=se)

plt.figure(figsize=(9,5))
plt.plot(x, pdf, lw=2)
plt.fill_between(x, pdf, where=(x >= ci_inferior) & (x <= ci_superior), alpha=0.3)
plt.axvline(x_bar, color='black', linestyle='--', label=f'Media muestral = {x_bar:.2f} cm')
plt.axvline(ci_inferior, color='red', linestyle='--', label=f'CI inferior = {ci_inferior:.2f} cm')
plt.axvline(ci_superior, color='red', linestyle='--', label=f'CI superior = {ci_superior:.2f} cm')
# Añadir anotaciones de Z críticos respecto a la media muestral:
plt.annotate(f'Z_crit = {z_crit:.3f}', xy=(x_bar + z_crit*se, stats.norm.pdf(x_bar + z_crit*se, loc=x_bar, scale=se)),
             xytext=(x_bar + z_crit*se + 0.2*se, max(pdf)*0.6), arrowprops=dict(arrowstyle='->'))
plt.title('Distribución muestral de la media (x̄) y CI 95%')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# --------------------------
# 5) Conclusión práctica (texto claro)
print("\nConclusión práctica:")
print(f"El intervalo de confianza del 95% para la altura promedio es [{ci_inferior:.2f} cm, {ci_superior:.2f} cm].")
print("Esto significa que, si repitiésemos el muestreo muchas veces y construyéramos el IC de la misma forma,")
print("aproximadamente el 95% de esos intervalos contendrían la verdadera media poblacional de la altura.")
```

```python
# Punto 6 - Intervalo de confianza del 95% para la media (σ desconocida, uso t de Student)
# Enunciado: n = 12, x̄ = 84 °C, s = 5 °C. Nivel de confianza = 95%.

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# --------------------------
# 1) Datos del problema
n = 12                   # tamaño de la muestra
x_bar = 84.0             # media muestral (°C)
s = 5.0                  # desviación estándar muestral (σ desconocida)
confianza = 0.95         # nivel de confianza

# --------------------------
# 2) Cálculos básicos (t-student)
alpha = 1 - confianza
df = n - 1                                   # grados de libertad
se = s / np.sqrt(n)                          # error estándar de la media (usando s)
t_crit = stats.t.ppf(1 - alpha/2, df=df)     # valor crítico t (bilateral)
margen = t_crit * se                         # margen de error
ci_inferior = x_bar - margen
ci_superior = x_bar + margen

# --------------------------
# 3) Impresión paso a paso
print("PUNTO 6 - INTERVALO DE CONFIANZA (t de Student, σ desconocida)")
print("--------------------------------------------------------------")
print(f"Datos: n = {n}, x̄ = {x_bar:.2f} °C, s = {s:.2f} °C")
print(f"Nivel de confianza = {confianza*100:.0f}%  →  α = {alpha:.3f}")
print(f"Grados de libertad = df = n - 1 = {df}")
print(f"Error estándar (SE) = s / √n = {se:.4f} °C")
print(f"Valor crítico t (bilateral, df={df}) = {t_crit:.4f}")
print(f"Margen de error = t_crit * SE = {margen:.4f} °C")
print(f"Intervalo de confianza del {confianza*100:.0f}%: [{ci_inferior:.4f} °C, {ci_superior:.4f} °C]")
print("\nInterpretación: Con un 95% de confianza, la verdadera temperatura media del café se encuentra dentro de este intervalo.")

# --------------------------
# 4) Gráfica: distribución t centrada en x̄ con escala = SE, sombreando el IC
x_min = x_bar - 4*se
x_max = x_bar + 4*se
x = np.linspace(x_min, x_max, 500)
pdf = stats.t.pdf((x - x_bar) / se, df=df) / se  # alternativa: usar loc/scale param (equivalente abajo)

# (Usaremos la forma directa con loc y scale para mayor claridad:)
pdf = stats.t.pdf(x, df=df, loc=x_bar, scale=se)

plt.figure(figsize=(9,5))
plt.plot(x, pdf, lw=2, label=f'Densidad t (df={df}) de la media muestral')
# sombrear el intervalo de confianza
plt.fill_between(x, pdf, where=(x >= ci_inferior) & (x <= ci_superior), alpha=0.3, label=f'IC {int(confianza*100)}%')
# líneas verticales importantes
plt.axvline(x_bar, color='black', linestyle='--', label=f'Media muestral = {x_bar:.2f} °C')
plt.axvline(ci_inferior, color='red', linestyle='--', label=f'CI inferior = {ci_inferior:.2f} °C')
plt.axvline(ci_superior, color='red', linestyle='--', label=f'CI superior = {ci_superior:.2f} °C')
# anotar valores críticos (en unidades de temperatura)
left_crit = x_bar - t_crit*se
right_crit = x_bar + t_crit*se
plt.annotate(f't_crit = {t_crit:.3f}\n(x = {left_crit:.2f})',
             xy=(left_crit, stats.t.pdf(left_crit, df=df, loc=x_bar, scale=se)),
             xytext=(left_crit - 0.8*se, max(pdf)*0.6),
             arrowprops=dict(arrowstyle='->'))
plt.annotate(f'(x = {right_crit:.2f})',
             xy=(right_crit, stats.t.pdf(right_crit, df=df, loc=x_bar, scale=se)),
             xytext=(right_crit + 0.05, max(pdf)*0.6),
             arrowprops=dict(arrowstyle='->'))

plt.title('Distribución t de la media muestral y IC 95% - Punto 6')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Densidad')
plt.grid(alpha=0.3)
plt.legend()
plt.show()

# --------------------------
# 5) Conclusión práctica (texto claro)
print("\nConclusión práctica:")
print(f"El intervalo de confianza del 95% para la temperatura media del café es [{ci_inferior:.2f} °C, {ci_superior:.2f} °C].")
print("Esto significa que, con el método y nivel de confianza empleados, estimamos que la verdadera temperatura media")
print("del café servido por la cadena se encuentra dentro de ese rango.")
```

```python
# Punto 10 - Intervalo de confianza del 95% para la media (t de Student, σ desconocida)
# Enunciado: n = 8, x̄ = 1.5 kg, s = 0.3 kg. Nivel de confianza = 95%.

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# --------------------------
# 1) Datos del problema
n = 8                   # tamaño de la muestra
x_bar = 1.5             # media muestral (kg)
s = 0.3                 # desviación estándar muestral (kg), σ poblacional desconocida
confianza = 0.95        # nivel de confianza

# --------------------------
# 2) Cálculos (uso de t de Student)
alpha = 1 - confianza
df = n - 1                                  # grados de libertad
se = s / np.sqrt(n)                         # error estándar de la media
t_crit = stats.t.ppf(1 - alpha/2, df=df)    # t crítico bilateral
margen = t_crit * se                        # margen de error
ci_inferior = x_bar - margen
ci_superior = x_bar + margen

# --------------------------
# 3) Salida paso a paso (clara y comentada)
print("PUNTO 10 - INTERVALO DE CONFIANZA (t de Student, σ desconocida)")
print("--------------------------------------------------------------")
print(f"Datos: n = {n}, x̄ = {x_bar:.3f} kg, s = {s:.3f} kg")
print(f"Nivel de confianza = {confianza*100:.0f}%  →  α = {alpha:.3f}")
print(f"Grados de libertad = df = n - 1 = {df}")
print(f"Error estándar (SE) = s / √n = {se:.4f} kg")
print(f"Valor crítico t (bilateral, df={df}) = {t_crit:.4f}")
print(f"Margen de error = t_crit * SE = {margen:.4f} kg")
print(f"Intervalo de confianza del {confianza*100:.0f}%: [{ci_inferior:.4f} kg, {ci_superior:.4f} kg]")
print("\nInterpretación: Con un 95% de confianza, la verdadera media del peso de libros de matemáticas")
print("se encuentra dentro de este intervalo. (n pequeño → usamos t de Student)")

# --------------------------
# 4) Gráfica: densidad t centrada en x̄ (escala = SE), sombreando el IC
x_min = x_bar - 4*se
x_max = x_bar + 4*se
x = np.linspace(x_min, x_max, 500)
pdf = stats.t.pdf(x, df=df, loc=x_bar, scale=se)

plt.figure(figsize=(9,5))
plt.plot(x, pdf, lw=2, label=f'Densidad t (df={df}) de la media muestral')
plt.fill_between(x, pdf, where=(x >= ci_inferior) & (x <= ci_superior), alpha=0.3, label=f'IC {int(confianza*100)}%')
# líneas verticales: media y límites del IC
plt.axvline(x_bar, color='black', linestyle='--', label=f'Media muestral = {x_bar:.3f} kg')
plt.axvline(ci_inferior, color='red', linestyle='--', label=f'CI inferior = {ci_inferior:.3f} kg')
plt.axvline(ci_superior, color='red', linestyle='--', label=f'CI superior = {ci_superior:.3f} kg')

# anotar t crítico en términos de x (x_bar ± t_crit*SE)
left_crit = x_bar - t_crit*se
right_crit = x_bar + t_crit*se
plt.annotate(f't_crit = {t_crit:.3f}\n(x = {left_crit:.3f} kg)',
             xy=(left_crit, stats.t.pdf(left_crit, df=df, loc=x_bar, scale=se)),
             xytext=(left_crit - 0.03, max(pdf)*0.65), arrowprops=dict(arrowstyle='->'))
plt.annotate(f'(x = {right_crit:.3f} kg)',
             xy=(right_crit, stats.t.pdf(right_crit, df=df, loc=x_bar, scale=se)),
             xytext=(right_crit + 0.005, max(pdf)*0.65), arrowprops=dict(arrowstyle='->'))

plt.title('Distribución t de la media muestral y IC 95% - Punto 10')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad')
plt.grid(alpha=0.3)
plt.legend()
plt.show()

# --------------------------
# 5) Conclusión práctica final
print("\nConclusión práctica final:")
print(f"El intervalo de confianza del 95% para el peso medio es [{ci_inferior:.3f} kg, {ci_superior:.3f} kg].")
print("Nota: con n=8 (muestra pequeña) y σ desconocida, el intervalo es más ancho que si tuviéramos σ conocida.")
```

```python
# Punto 11 - Tamaño de muestra para proporción (IC 95%, margen de error 3%)
# Enunciado: estudio piloto p̂ = 0.72. Calcular n para E = 0.03 y confianza 95%.
# Además: ¿qué n si no tuviéramos estudio piloto? (usar p = 0.5 conservadoramente)

import math
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# --------------------------
# 1) Datos del problema
p_hat = 0.72          # proporción observada en el estudio piloto
E = 0.03              # margen de error deseado (3%)
confianza = 0.95      # nivel de confianza

# --------------------------
# 2) Cálculo del valor crítico z para intervalo bilateral
alpha = 1 - confianza
z_crit = stats.norm.ppf(1 - alpha/2)   # z_{1-α/2}

# --------------------------
# 3) Fórmula para tamaño de muestra (proporción)
# n = (z^2 * p * (1-p)) / E^2
n_pilot = (z_crit**2 * p_hat * (1 - p_hat)) / (E**2)
n_pilot_ceil = math.ceil(n_pilot)      # siempre redondear hacia arriba

# Si no hay estudio piloto, usar p = 0.5 (maximiza la varianza)
p_conservador = 0.5
n_conservative = (z_crit**2 * p_conservador * (1 - p_conservador)) / (E**2)
n_conservative_ceil = math.ceil(n_conservative)

# --------------------------
# 4) Imprimir resultados y explicación
print("PUNTO 11 - TAMAÑO DE MUESTRA PARA PROPORCIÓN (IC 95%, E = 3%)")
print("----------------------------------------------------------------")
print(f"Nivel de confianza = {confianza*100:.0f}%  →  α = {alpha:.3f}, z_crit = {z_crit:.4f}")
print(f"Margen de error deseado E = {E:.3f} (3%)")
print("\nCon estudio piloto (p̂ = {:.2f}):".format(p_hat))
print(f"  Cálculo continuo n = {n_pilot:.4f}")
print(f"  Tamaño de muestra redondeado hacia arriba n = {n_pilot_ceil}")
print("\nSi NO hay estudio piloto (usar p = 0.5 conservador):")
print(f"  Cálculo continuo n = {n_conservative:.4f}")
print(f"  Tamaño de muestra redondeado hacia arriba n = {n_conservative_ceil}")
print("\nObservación:")
print(" - Se redondea siempre hacia arriba para garantizar el margen de error requerido.")
print(" - Si la población es pequeña y conocida, aplicar corrección por población finita (FPC).")

# --------------------------
# 5) Gráfica (opcional pero útil): distribución muestral de la proporción usando n calculado con p̂
# Visualizamos la aproximación normal de p̂ con n = n_pilot_ceil
n_plot = n_pilot_ceil
mean_p = p_hat
se_p = np.sqrt(mean_p * (1 - mean_p) / n_plot)

x = np.linspace(max(0, mean_p - 6*se_p), min(1, mean_p + 6*se_p), 500)
pdf = stats.norm.pdf(x, loc=mean_p, scale=se_p)

plt.figure(figsize=(9,5))
plt.plot(x, pdf, lw=2, label=f'Approx. normal de p̂ (n={n_plot})')
# sombrear la región dentro del margen E alrededor de p_hat
plt.fill_between(x, pdf, where=(x >= mean_p - E) & (x <= mean_p + E), alpha=0.25,
                 label=f'±E = ±{E:.3f}')
plt.axvline(mean_p, color='black', linestyle='--', label=f'p̂ = {mean_p:.2f}')
plt.axvline(mean_p - E, color='red', linestyle='--', label=f'p̂ - E = {mean_p - E:.3f}')
plt.axvline(mean_p + E, color='red', linestyle='--', label=f'p̂ + E = {mean_p + E:.3f}')
plt.title('Aproximación normal de la distribución muestral de la proporción (usando n calculado)')
plt.xlabel('Proporción')
plt.ylabel('Densidad')
plt.grid(alpha=0.3)
plt.legend()
plt.show()

# --------------------------
# 6) Conclusión práctica corta
print("\nConclusión práctica:")
print(f" - Con p̂ = {p_hat:.2f}, confianza 95% y E = 3%, se requiere una muestra de **n = {n_pilot_ceil}**.")
print(f" - Si no hubiera estudio piloto (usar p = 0.5), la muestra requerida sería **n = {n_conservative_ceil}** (más conservadora).")
```

```python
# Punto 21 - Prueba de hipótesis para una media (σ conocida, Z-test)
# Enunciado: μ0 = 50, σ = 10, n = 36, x̄ = 53. Nivel de significancia α = 0.05.
# Hipótesis: H0: μ = 50 vs H1: μ ≠ 50 (prueba bilateral).

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# --------------------------
# 1) Datos del problema
mu0 = 50            # media poblacional bajo H0
sigma = 10          # desviación estándar poblacional (conocida)
n = 36              # tamaño de muestra
x_bar = 53          # media muestral
alpha = 0.05        # nivel de significancia
confianza = 1 - alpha

# --------------------------
# 2) Estadístico de prueba (Z)
se = sigma / np.sqrt(n)            # error estándar
z_stat = (x_bar - mu0) / se        # estadístico Z
z_crit = stats.norm.ppf(1 - alpha/2)  # valor crítico bilateral

# p-valor (bilateral)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# --------------------------
# 3) Imprimir resultados
print("PUNTO 21 - PRUEBA DE HIPÓTESIS PARA UNA MEDIA (σ conocida)")
print("---------------------------------------------------------")
print(f"H0: μ = {mu0}")
print(f"H1: μ ≠ {mu0}  (prueba bilateral)")
print(f"Datos: n = {n}, x̄ = {x_bar}, σ = {sigma}, α = {alpha}")
print(f"Error estándar (SE) = {se:.4f}")
print(f"Estadístico de prueba Z = {z_stat:.4f}")
print(f"Valor crítico Z (±z_{1-alpha/2}) = ±{z_crit:.4f}")
print(f"p-valor (bilateral) = {p_value:.6f}")

# Decisión
if abs(z_stat) > z_crit:
    print("Decisión: |Z| > Z crítico → Rechazamos H0.")
else:
    print("Decisión: |Z| ≤ Z crítico → No se rechaza H0.")

# Interpretación
if abs(z_stat) > z_crit:
    print(f"Conclusión: Existe evidencia significativa (α = {alpha}) de que la media poblacional difiere de {mu0}.")
else:
    print(f"Conclusión: No hay evidencia suficiente (α = {alpha}) para afirmar que la media poblacional difiere de {mu0}.")

# --------------------------
# 4) Gráfica ilustrativa
x = np.linspace(mu0 - 5*se, mu0 + 5*se, 500)
pdf = stats.norm.pdf(x, mu0, se)

plt.figure(figsize=(9,5))
plt.plot(x, pdf, lw=2, label='Distribución bajo H0 (μ=50)')
# regiones críticas
plt.fill_between(x, pdf, where=(x <= mu0 - z_crit*se), color='red', alpha=0.3, label='Región crítica izquierda')
plt.fill_between(x, pdf, where=(x >= mu0 + z_crit*se), color='red', alpha=0.3, label='Región crítica derecha')
# valor observado
plt.axvline(x_bar, color='blue', linestyle='--', lw=2, label=f'x̄ observada = {x_bar}')
# medias y límites críticos
plt.axvline(mu0, color='black', linestyle='--', label='μ0 = 50')
plt.axvline(mu0 - z_crit*se, color='red', linestyle='--', label=f'Límite crítico inferior = {mu0 - z_crit*se:.2f}')
plt.axvline(mu0 + z_crit*se, color='red', linestyle='--', label=f'Límite crítico superior = {mu0 + z_crit*se:.2f}')

plt.title('Prueba de hipótesis para la media (σ conocida) - Punto 21')
plt.xlabel('Media muestral posible')
plt.ylabel('Densidad')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Punto 22 - Prueba de hipótesis (Z-test, σ conocida) - Prueba unilateral a la izquierda
# Enunciado: fabricante afirma μ = 1200 horas. Cliente alega que dura menos.
# Muestra: n = 50, x̄ = 1180, σ = 100. Nivel de significancia α = 0.01.
# H0: μ = 1200    vs    H1: μ < 1200

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# --------------------------
# 1) Datos del problema
mu0 = 1200.0       # media bajo H0
x_bar = 1180.0     # media muestral
sigma = 100.0      # desviación estándar poblacional conocida
n = 50             # tamaño de la muestra
alpha = 0.01       # nivel de significancia (unilateral izquierda)

# --------------------------
# 2) Estadístico de prueba Z
se = sigma / np.sqrt(n)                 # error estándar de la media
z_stat = (x_bar - mu0) / se             # estadístico Z (observado)

# Para prueba unilateral a la izquierda:
z_crit = stats.norm.ppf(alpha)          # valor crítico (cola izquierda)
p_value_one_sided = stats.norm.cdf(z_stat)   # p-valor para H1: μ < μ0

# --------------------------
# 3) Imprimir resultados paso a paso
print("PUNTO 22 - PRUEBA Z UNILATERAL (H1: μ < 1200)")
print("------------------------------------------------")
print(f"H0: μ = {mu0}")
print(f"H1: μ < {mu0}  (prueba a la izquierda)")
print(f"Datos: n = {n}, x̄ = {x_bar}, σ = {sigma}, α = {alpha}")
print(f"Error estándar (SE) = σ / √n = {se:.4f}")
print(f"Estadístico de prueba Z = (x̄ - μ0) / SE = {z_stat:.4f}")
print(f"Valor crítico (cola izquierda) z_crit = {z_crit:.4f}")
print(f"p-valor (unilateral, cola izquierda) = P(Z ≤ {z_stat:.4f}) = {p_value_one_sided:.6f}")

# Decisión
if z_stat <= z_crit:
    print("\nDecisión: Z ≤ z_crit → Rechazamos H0 (a nivel α).")
else:
    print("\nDecisión: Z > z_crit → No rechazamos H0 (a nivel α).")

# Interpretación práctica
if p_value_one_sided < alpha:
    print(f"Conclusión: Hay evidencia estadística (α={alpha}) de que la duración media es menor que {mu0} horas.")
else:
    print(f"Conclusión: No hay evidencia suficiente (α={alpha}) para afirmar que la duración media sea menor que {mu0} horas.")

# --------------------------
# 4) Gráfica ilustrativa: distribución de la media bajo H0 (normal con mean=mu0, scale=SE)
x = np.linspace(mu0 - 5*se, mu0 + 5*se, 800)
pdf = stats.norm.pdf(x, loc=mu0, scale=se)

plt.figure(figsize=(10,5))
plt.plot(x, pdf, lw=2, label='Distribución de x̄ bajo H0 (N(μ0, SE))')
# Región de rechazo (cola izquierda)
x_reject = np.linspace(mu0 - 5*se, mu0 + z_crit*se, 400)
plt.fill_between(x_reject, stats.norm.pdf(x_reject, loc=mu0, scale=se), color='red', alpha=0.3, label='Región de rechazo (α)')
# Marcar la media muestral observada
plt.axvline(x_bar, color='blue', linestyle='--', lw=2, label=f'Media muestral x̄ = {x_bar:.1f}')
# Marcar el valor crítico en unidades de x̄
x_crit_value = mu0 + z_crit*se
plt.axvline(x_crit_value, color='red', linestyle='--', lw=2, label=f'Valor crítico x = {x_crit_value:.2f} (z_crit={z_crit:.3f})')

plt.title('Prueba Z unilateral (cola izquierda) - Punto 22')
plt.xlabel('Valor de la media muestral (horas)')
plt.ylabel('Densidad')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Punto 23 - Prueba de hipótesis para una media (σ conocida, Z-test, bilateral)
# Enunciado: Profesor compara su clase con promedio nacional μ0 = 75.
# Muestra: n = 36, x̄ = 78, σ = 12. Nivel de significancia α = 0.05.
# Hipótesis: H0: μ = 75  vs  H1: μ ≠ 75  (prueba bilateral)

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# --------------------------
# 1) Datos del problema
mu0 = 75.0        # media bajo H0
x_bar = 78.0      # media muestral
sigma = 12.0      # desviación estándar poblacional (conocida)
n = 36            # tamaño de muestra
alpha = 0.05      # nivel de significancia (bilateral)

# --------------------------
# 2) Estadístico de prueba (Z) y valores críticos
se = sigma / np.sqrt(n)                 # error estándar de la media
z_stat = (x_bar - mu0) / se             # estadístico Z observado
z_crit = stats.norm.ppf(1 - alpha/2)    # valor crítico positivo (bilateral)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # p-valor bilateral

# --------------------------
# 3) Impresión paso a paso
print("PUNTO 23 - PRUEBA DE HIPÓTESIS PARA UNA MEDIA (σ conocida)")
print("----------------------------------------------------------")
print(f"H0: μ = {mu0}")
print(f"H1: μ ≠ {mu0}  (prueba bilateral)")
print(f"Datos: n = {n}, x̄ = {x_bar}, σ = {sigma}, α = {alpha}")
print(f"Error estándar (SE) = σ / √n = {se:.4f}")
print(f"Estadístico de prueba Z = (x̄ - μ0) / SE = {z_stat:.4f}")
print(f"Valor crítico bilateral ±Z = ±{z_crit:.4f}")
print(f"p-valor (bilateral) = {p_value:.6f}")

# Decisión
if abs(z_stat) > z_crit:
    decision_text = "Rechazamos H0"
else:
    decision_text = "No rechazamos H0"

print(f"Decisión (α = {alpha}): {decision_text}")

# Interpretación práctica
if p_value < alpha:
    print(f"Conclusión: Hay evidencia estadística significativa (α={alpha}) de que la media de la clase difiere de {mu0}.")
else:
    print(f"Conclusión: No hay evidencia suficiente (α={alpha}) para afirmar que la media de la clase difiera de {mu0}.")

# --------------------------
# 4) Gráfica ilustrativa: distribución de x̄ bajo H0 y regiones de rechazo
x = np.linspace(mu0 - 5*se, mu0 + 5*se, 800)
pdf = stats.norm.pdf(x, loc=mu0, scale=se)

plt.figure(figsize=(10,5))
plt.plot(x, pdf, lw=2, label='Distribución de x̄ under H0 (N(μ0, SE))')

# regiones críticas (dos colas)
left_limit = mu0 - z_crit*se
right_limit = mu0 + z_crit*se
plt.fill_between(x, pdf, where=(x <= left_limit), color='red', alpha=0.3, label='Región crítica (cola izquierda)')
plt.fill_between(x, pdf, where=(x >= right_limit), color='red', alpha=0.3, label='Región crítica (cola derecha)')

# marcar media muestral observada y μ0
plt.axvline(x_bar, color='blue', linestyle='--', lw=2, label=f'x̄ observado = {x_bar}')
plt.axvline(mu0, color='black', linestyle='--', label=f'μ0 = {mu0}')
plt.axvline(left_limit, color='red', linestyle='--', label=f'Límite crítico inferior = {left_limit:.2f}')
plt.axvline(right_limit, color='red', linestyle='--', label=f'Límite crítico superior = {right_limit:.2f}')

plt.title('Prueba Z bilateral para la media - Punto 23')
plt.xlabel('Valor de la media muestral (x̄)')
plt.ylabel('Densidad')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# --------------------------
# 5) Resumen final (numérico claro)
print("\nResumen numérico:")
print(f"  SE = {se:.4f}")
print(f"  Z estadístico = {z_stat:.4f}")
print(f"  Z crítico (±) = ±{z_crit:.4f}")
print(f"  p-valor bilateral = {p_value:.6f}")
print(f"  Decisión: {decision_text}")
```

```python
# Punto 24 - Prueba de hipótesis para una media (σ conocida, Z-test)
# Enunciado: El laboratorio afirma que su medicamento reduce la presión arterial en 10 mmHg en promedio.
#           Estudio: n = 25, x̄ = 8 mmHg (reducción promedio observada), σ = 3 mmHg. α = 0.05.
# Objetivo: seguir el protocolo (H0/H1, α, estadístico, valor crítico, p-valor, gráfica, conclusión).
# Hipótesis (prueba bilateral por defecto):
#   H0: μ = 10    (la reducción media es 10 mmHg)
#   H1: μ ≠ 10    (la reducción media difiere de 10 mmHg)
# Nota: también se muestra el p-valor unilateral (cola izquierda) por si se desea interpretar como prueba H1: μ < 10.

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# --------------------------
# 1) Datos del problema
mu0 = 10.0        # reducción declarada bajo H0 (mmHg)
x_bar = 8.0       # reducción observada (mmHg)
sigma = 3.0       # desviación estándar poblacional conocida (mmHg)
n = 25            # tamaño de la muestra
alpha = 0.05      # nivel de significancia (bilateral)

# --------------------------
# 2) Estadístico de prueba (Z) y parámetros
se = sigma / np.sqrt(n)               # error estándar de la media
z_stat = (x_bar - mu0) / se           # estadístico Z observado
z_crit = stats.norm.ppf(1 - alpha/2)  # valor crítico positivo para prueba bilateral

# p-valores
p_value_bilateral = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # p bilateral
p_value_left = stats.norm.cdf(z_stat)                     # p unilateral (cola izquierda)
p_value_right = 1 - stats.norm.cdf(z_stat)                # p unilateral (cola derecha)

# --------------------------
# 3) Impresión paso a paso
print("PUNTO 24 - PRUEBA DE HIPÓTESIS PARA UNA MEDIA (σ conocida)")
print("-----------------------------------------------------------")
print(f"H0: μ = {mu0} mmHg")
print(f"H1: μ ≠ {mu0} mmHg  (prueba bilateral por defecto)")
print(f"Datos: n = {n}, x̄ = {x_bar:.2f} mmHg, σ = {sigma:.2f} mmHg, α = {alpha}")
print(f"Error estándar (SE) = σ / √n = {se:.4f} mmHg")
print(f"Estadístico Z = (x̄ - μ0) / SE = {z_stat:.4f}")
print(f"Valor crítico bilateral ±Z = ±{z_crit:.4f}")
print(f"p-valor (bilateral) = {p_value_bilateral:.6f}")
print(f"p-valor (unilateral, cola izquierda) = {p_value_left:.6f}")
print()

# Decisión según prueba bilateral
if abs(z_stat) > z_crit:
    print("Decisión (bilateral): |Z| > Z_crit  → Rechazamos H0 al nivel α.")
else:
    print("Decisión (bilateral): |Z| ≤ Z_crit  → No se rechaza H0 al nivel α.")

# Interpretación práctica
if p_value_bilateral < alpha:
    print(f"Conclusión (bilateral): Hay evidencia estadística significativa (α={alpha}) de que")
    print("la reducción promedio difiere de 10 mmHg. En la muestra observamos una reducción promedio de 8 mmHg,")
    print("lo cual es significativamente distinto del valor declarado (10 mmHg).")
else:
    print(f"Conclusión (bilateral): No hay evidencia suficiente (α={alpha}) para afirmar que la reducción difiera de 10 mmHg.")

# Si el interés real fuera demostrar que la reducción es menor que 10 (H1: μ < 10), usar p_value_left:
print()
print("Nota sobre prueba unilateral (H1: μ < 10):")
if p_value_left < alpha:
    print(f"  - p-valor (cola izquierda) = {p_value_left:.6f} < {alpha}  → Rechazamos H0 en favor de H1: μ < 10.")
    print("    Interpretación: hay evidencia de que la reducción media es MENOR que 10 mmHg.")
else:
    print(f"  - p-valor (cola izquierda) = {p_value_left:.6f} ≥ {alpha}  → No rechazamos H0 en la prueba unilateral.")

# --------------------------
# 4) Gráfica de la distribución bajo H0 (normal) indicando regiones de rechazo, valor crítico y estadístico observado
x = np.linspace(mu0 - 5*se, mu0 + 5*se, 800)
pdf = stats.norm.pdf(x, loc=mu0, scale=se)

plt.figure(figsize=(10,5))
plt.plot(x, pdf, lw=2, label=f'Distribución de x̄ bajo H0: N(μ0={mu0}, SE={se:.3f})')

# Regiones críticas (dos colas)
left_limit = mu0 - z_crit*se
right_limit = mu0 + z_crit*se
plt.fill_between(x, pdf, where=(x <= left_limit), color='red', alpha=0.3, label='Región de rechazo (cola izquierda)')
plt.fill_between(x, pdf, where=(x >= right_limit), color='red', alpha=0.3, label='Región de rechazo (cola derecha)')

# Marcar media observada x̄ y límites
plt.axvline(x_bar, color='blue', linestyle='--', lw=2, label=f'x̄ observado = {x_bar:.2f} mmHg')
plt.axvline(mu0, color='black', linestyle='--', label=f'μ0 = {mu0} mmHg')
plt.axvline(left_limit, color='red', linestyle='--', label=f'Límite crítico inferior = {left_limit:.2f}')
plt.axvline(right_limit, color='red', linestyle='--', label=f'Límite crítico superior = {right_limit:.2f}')

# Añadir texto con el valor Z en la gráfica
plt.text(x_bar, max(pdf)*0.8, f'Z_obs = {z_stat:.3f}', color='blue', ha='center')

plt.title('Prueba Z bilateral para la media - Punto 24')
plt.xlabel('Reducción media observada (mmHg)')
plt.ylabel('Densidad')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Punto 25 - Prueba de hipótesis para una media (σ conocida, Z-test, bilateral)
# Enunciado: Máquina llena botellas de 500 mL. Muestra: n = 60, x̄ = 498 mL, σ = 5 mL. α = 0.01.
# Objetivo: seguir el protocolo: H0/H1, α, estadístico (Z), valor crítico y región de rechazo, p-valor, gráfica y conclusión.

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# --------------------------
# 1) Datos y formulación de hipótesis
mu0 = 500.0       # valor declarado por la máquina (mL)
x_bar = 498.0     # media muestral (mL)
sigma = 5.0       # desviación estándar poblacional conocida (mL)
n = 60            # tamaño de la muestra
alpha = 0.01      # nivel de significancia

# Hipótesis (prueba bilateral: se sospecha que la máquina no llena exactamente 500 mL)
# H0: μ = mu0
# H1: μ ≠ mu0

# --------------------------
# 2) Estadístico de prueba Z
se = sigma / np.sqrt(n)                # error estándar de la media
z_stat = (x_bar - mu0) / se            # estadístico Z observado

# --------------------------
# 3) Valor crítico y región de rechazo (bilateral)
z_crit_pos = stats.norm.ppf(1 - alpha/2)   # positivo
z_crit_neg = -z_crit_pos                   # negativo

# --------------------------
# 4) p-valor (bilateral)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# --------------------------
# 5) Impresión paso a paso (resultados claros)
print("PUNTO 25 - PRUEBA Z PARA UNA MEDIA (σ conocida) - Máquina llenado 500 mL")
print("---------------------------------------------------------------------")
print(f"H0: μ = {mu0} mL")
print("H1: μ ≠ {0:.1f} mL".format(mu0))
print(f"Datos: n = {n}, x̄ = {x_bar:.2f} mL, σ = {sigma:.2f} mL, α = {alpha}")
print(f"Error estándar (SE) = σ / √n = {se:.4f} mL")
print(f"Estadístico Z observado = (x̄ - μ0) / SE = {z_stat:.4f}")
print(f"Valores críticos (bilateral) = {z_crit_neg:.4f}, {z_crit_pos:.4f}")
print(f"p-valor (bilateral) = {p_value:.6f}")

# Decisión basada en región crítica
if (z_stat <= z_crit_neg) or (z_stat >= z_crit_pos):
    decision = "Rechazamos H0"
else:
    decision = "No se rechaza H0"

print(f"Decisión (región crítica, α={alpha}): {decision}")

# --------------------------
# 6) Gráfica: distribución de x̄ bajo H0 (normal con mean=mu0, scale=SE), sombrear regiones de rechazo
x = np.linspace(mu0 - 5*se, mu0 + 5*se, 800)
pdf = stats.norm.pdf(x, loc=mu0, scale=se)

plt.figure(figsize=(10,5))
plt.plot(x, pdf, lw=2, label=f'Distribución de x̄ bajo H0: N(μ0={mu0}, SE={se:.3f})')

# regiones de rechazo (dos colas)
left_reject_x = x <= (mu0 + z_crit_neg*se)
right_reject_x = x >= (mu0 + z_crit_pos*se)
plt.fill_between(x, pdf, where=left_reject_x, color='red', alpha=0.3, label='Región de rechazo (cola izquierda)')
plt.fill_between(x, pdf, where=right_reject_x, color='red', alpha=0.3, label='Región de rechazo (cola derecha)')

# marcar estadístico observado y límites críticos en unidades de x̄
plt.axvline(x_bar, color='blue', linestyle='--', lw=2, label=f'x̄ observado = {x_bar} mL (Z = {z_stat:.3f})')
x_crit_neg = mu0 + z_crit_neg*se
x_crit_pos = mu0 + z_crit_pos*se
plt.axvline(x_crit_neg, color='red', linestyle='--', lw=1.5, label=f'Crítico inferior = {x_crit_neg:.2f} mL (z={z_crit_neg:.3f})')
plt.axvline(x_crit_pos, color='red', linestyle='--', lw=1.5, label=f'Crítico superior = {x_crit_pos:.2f} mL (z={z_crit_pos:.3f})')

plt.title('Prueba Z bilateral para la media - Punto 25')
plt.xlabel('Valor de la media muestral (mL)')
plt.ylabel('Densidad')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# --------------------------
# 7) Conclusión práctica final
print("\nConclusión práctica:")
if decision.startswith("Rechazamos"):
    print(f"Con α = {alpha}, |Z_obs| = {abs(z_stat):.3f} > Z_crit = {z_crit_pos:.3f}, por lo tanto RECHAZAMOS H0.")
    print(f"Interpretación: hay evidencia estadística significativa de que la máquina NO llena en promedio 500 mL.")
    print(f"En la muestra observada la media fue {x_bar} mL, lo que sugiere que la máquina está subllenando (promedio menor).")
    print(f"p-valor bilateral = {p_value:.6f} (muy por debajo de α = {alpha}) → resultado estadísticamente significativo.")
else:
    print(f"Con α = {alpha}, |Z_obs| = {abs(z_stat):.3f} ≤ Z_crit = {z_crit_pos:.3f}, por lo tanto NO rechazamos H0.")
    print("Interpretación: no hay evidencia suficiente para afirmar que la máquina no esté llenando 500 mL en promedio.")
```

```python
# Punto 26 - Prueba t unilateral (cola derecha): ¿el tiempo promedio de entrega es MAYOR que 25 min?
# Enunciado: n = 15, x̄ = 28 min, s = 4 min, α = 0.05. H0: μ = 25 vs H1: μ > 25

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# --------------------------
# 1) Datos del problema
n = 15                      # tamaño de la muestra
x_bar = 28.0                # media muestral (minutos)
s = 4.0                     # desviación estándar muestral (minutos)
mu0 = 25.0                  # valor bajo H0 (minutos)
alpha = 0.05                # nivel de significancia (prueba unilateral derecha)

# --------------------------
# 2) Cálculos del estadístico t
df = n - 1
se = s / np.sqrt(n)                          # error estándar de la media
t_stat = (x_bar - mu0) / se                  # estadístico t observado
t_crit = stats.t.ppf(1 - alpha, df=df)      # valor crítico (unilateral derecha)
p_value_one_sided = 1 - stats.t.cdf(t_stat, df=df)  # p-valor (cola derecha)

# --------------------------
# 3) Impresión paso a paso (clara y comentada)
print("PUNTO 26 - PRUEBA t (unilateral derecha) - Tiempo de entrega")
print("------------------------------------------------------------")
print(f"H0: μ = {mu0} minutos")
print(f"H1: μ > {mu0} minutos (cliente sospecha que es mayor)")
print(f"Datos: n = {n}, x̄ = {x_bar:.2f} min, s = {s:.2f} min, α = {alpha}")
print(f"Grados de libertad = df = {df}")
print(f"Error estándar (SE) = s / √n = {se:.4f} min")
print(f"Estadístico t observado = (x̄ - μ0) / SE = {t_stat:.6f}")
print(f"Valor crítico t (unilateral, 1-α) = {t_crit:.6f}")
print(f"P-valor (unilateral, cola derecha) = {p_value_one_sided:.6f}")

# Decisión según región crítica y p-valor
if t_stat > t_crit:
    decision = "Rechazamos H0"
else:
    decision = "No rechazamos H0"

print("\nDecisión basada en la regla de rechazo:")
print(f"  {decision} (porque t_obs {'>' if t_stat > t_crit else '<='} t_crit)")

print("\nDecisión basada en p-valor:")
if p_value_one_sided < alpha:
    print(f"  p-valor = {p_value_one_sided:.6f} < α = {alpha} → Rechazamos H0")
else:
    print(f"  p-valor = {p_value_one_sided:.6f} ≥ α = {alpha} → No rechazamos H0")

# Interpretación práctica
if (t_stat > t_crit) and (p_value_one_sided < alpha):
    print(f"\nConclusión práctica: Hay evidencia estadística (α = {alpha}) para afirmar que")
    print(f"el tiempo promedio de entrega es mayor que {mu0} minutos. (Media muestral = {x_bar} min)")
else:
    print(f"\nConclusión práctica: No hay evidencia suficiente para afirmar que el tiempo promedio")
    print(f"de entrega sea mayor que {mu0} minutos.")

# --------------------------
# 4) Gráfica: distribución t de la media bajo H0 (t con loc=mu0, scale=SE)
x_min = mu0 - 4*se
x_max = mu0 + 6*se
x = np.linspace(x_min, x_max, 800)
pdf = stats.t.pdf(x, df=df, loc=mu0, scale=se)

plt.figure(figsize=(10,5))
plt.plot(x, pdf, lw=2, label=f'Densidad de x̄ bajo H0: t(df={df}, loc={mu0}, scale=SE)')
# sombrear región de rechazo (cola derecha)
x_reject = x >= (mu0 + t_crit*se)
plt.fill_between(x, pdf, where=x_reject, color='red', alpha=0.3, label=f'Región de rechazo (α = {alpha})')
# marcar media muestral observada y el valor crítico en unidades de x̄
x_crit_value = mu0 + t_crit * se
plt.axvline(x_crit_value, color='red', linestyle='--', lw=1.5, label=f'Valor crítico x = {x_crit_value:.3f} (t_crit={t_crit:.3f})')
plt.axvline(x_bar, color='blue', linestyle='--', lw=2, label=f'x̄ observado = {x_bar:.3f} (t_obs={t_stat:.3f})')
plt.axvline(mu0, color='black', linestyle=':', label=f'μ0 = {mu0}')

plt.title('Prueba t (unilateral derecha) - Punto 26')
plt.xlabel('Valor de la media muestral (minutos)')
plt.ylabel('Densidad')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Punto 29 - Prueba t (n pequeña, σ desconocida)
# Enunciado: Suplemento promete aumento μ0 = 2 mg/dL.
# Muestra: n = 8, x̄ = 1.5 mg/dL, s = 0.6 mg/dL. Nivel de significancia α = 0.10.
# Objetivo: seguir el protocolo (plantear H0/H1, calcular t, p-valor, región crítica, gráfica y conclusión).
# Nota: muestro la prueba bilateral por defecto y la interpretación para la prueba unilateral a la izquierda
# (H1: μ < 2), que tiene sentido dado que la observación x̄ < μ0 (se comprueba si el suplemento logra el aumento prometido).

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# --------------------------
# 1) Datos
n = 8
x_bar = 1.5
s = 0.6
mu0 = 2.0
alpha = 0.10

# --------------------------
# 2) Estadístico t
df = n - 1
se = s / np.sqrt(n)
t_stat = (x_bar - mu0) / se

# p-valores
# bilateral: p = 2 * P(T >= |t_obs|)
p_two_sided = 2 * (1 - stats.t.cdf(abs(t_stat), df=df))
# unilateral izquierda (H1: μ < mu0): p = P(T <= t_obs)
p_left = stats.t.cdf(t_stat, df=df)
# unilateral derecha (H1: μ > mu0): p = 1 - P(T <= t_obs)
p_right = 1 - stats.t.cdf(t_stat, df=df)

# valores críticos
t_crit_bilateral = stats.t.ppf(1 - alpha/2, df=df)
t_crit_left = stats.t.ppf(alpha, df=df)        # valor crítico para cola izquierda
t_crit_right = stats.t.ppf(1 - alpha, df=df)   # valor crítico para cola derecha

# --------------------------
# 3) Impresión paso a paso
print("PUNTO 29 - PRUEBA t (σ desconocida)")
print("--------------------------------------------------")
print(f"H0: μ = {mu0}")
print("Opciones de alternativa (ver interpretación):")
print(" - Bilateral: H1: μ ≠ {0}".format(mu0))
print(" - Unilateral izquierda: H1: μ < {0}  (comprobar si el suplemento NO alcanza lo prometido)".format(mu0))
print()
print(f"Datos: n = {n}, x̄ = {x_bar:.3f}, s = {s:.3f}, α = {alpha}")
print(f"Grados de libertad: df = {df}")
print(f"Error estándar (SE) = s / √n = {se:.4f}")
print(f"Estadístico t observado = (x̄ - μ0) / SE = {t_stat:.4f}")
print()
print("---------- Prueba bilateral (H1: μ ≠ {0}) ----------".format(mu0))
print(f"Valor crítico (bilateral, ±t_{{1-α/2,df}}) = ±{t_crit_bilateral:.4f}")
print(f"p-valor (bilateral) = {p_two_sided:.6f}")
if abs(t_stat) > t_crit_bilateral:
    print(f"Decisión bilateral (α={alpha}): |t_obs| > t_crit → Rechazamos H0.")
else:
    print(f"Decisión bilateral (α={alpha}): |t_obs| ≤ t_crit → No rechazamos H0.")
print()

print("------ Prueba unilateral izquierda (H1: μ < {0}) ------".format(mu0))
print(f"Valor crítico (cola izquierda) t_crit = {t_crit_left:.4f}")
print(f"p-valor (cola izquierda) = P(T ≤ t_obs) = {p_left:.6f}")
if p_left < alpha:
    print(f"Decisión unilateral izquierda (α={alpha}): p < α → Rechazamos H0 en favor de H1: μ < {mu0}.")
else:
    print(f"Decisión unilateral izquierda (α={alpha}): p ≥ α → No rechazamos H0 (no hay evidencia de μ < {mu0}).")
print()

# --------------------------
# 4) Interpretación práctica resumida
print("Interpretación práctica (recomendación):")
if p_left < alpha:
    print(f" Con la prueba unilateral a la izquierda (más apropiada aquí porque queremos saber si el suplemento")
    print(f"  NO alcanza el aumento prometido), p = {p_left:.4f} < {alpha} → hay evidencia de que el aumento medio")
    print(f"  es MENOR que {mu0} mg/dL. (x̄ = {x_bar}, n = {n})")
else:
    print(f" Con α = {alpha} no hay evidencia suficiente de que el aumento medio sea menor que {mu0} mg/dL.")
print(" Nota: la prueba bilateral no rechaza H0 tampoco (p bilateral = {:.4f}).".format(p_two_sided))

# --------------------------
# 5) Gráfica: densidad t bajo H0 (centrada en μ0 con escala = SE). Marco regiones relevantes.
x_min = mu0 - 6*se
x_max = mu0 + 6*se
x = np.linspace(x_min, x_max, 800)
pdf = stats.t.pdf(x, df=df, loc=mu0, scale=se)

plt.figure(figsize=(10,5))
plt.plot(x, pdf, lw=2, label=f'Distribución de x̄ bajo H0: t(df={df}, loc={mu0:.2f}, scale=SE={se:.4f})')

# sombrear región rechazo unilateral izquierda (si se usa H1: μ < mu0)
x_reject_left = x <= (mu0 + t_crit_left*se)
plt.fill_between(x, pdf, where=x_reject_left, color='red', alpha=0.25, label=f'Región rechazo (cola izq, α={alpha})')

# marcar estadístico observado en unidades de x̄
plt.axvline(x_bar, color='blue', linestyle='--', lw=2, label=f'x̄ observado = {x_bar:.3f} (t = {t_stat:.3f})')

# marcar H0 y límites críticos (en unidades de x̄)
plt.axvline(mu0, color='black', linestyle='--', label=f'μ0 = {mu0}')
plt.axvline(mu0 + t_crit_left*se, color='red', linestyle='--', label=f'Crit. unilateral izq x = {mu0 + t_crit_left*se:.3f}')
plt.axvline(mu0 - t_crit_bilateral*se, color='orange', linestyle='--', label=f'Crit. bilateral inf x = {mu0 - t_crit_bilateral*se:.3f}')
plt.axvline(mu0 + t_crit_bilateral*se, color='orange', linestyle='--', label=f'Crit. bilateral sup x = {mu0 + t_crit_bilateral*se:.3f}')

plt.title('Prueba t para la media - Punto 29')
plt.xlabel('Aumento medio observado (mg/dL)')
plt.ylabel('Densidad')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Punto 30 - Prueba t bilateral (σ desconocida)
# Enunciado: Fabricante asegura vida útil μ0 = 10000 horas.
# Muestra: n = 20, x̄ = 9800 horas, s = 300 horas. Nivel de significancia α = 0.05.
# Hipótesis: H0: μ = 10000  vs  H1: μ ≠ 10000 (prueba bilateral).

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# --------------------------
# 1) Datos del problema
n = 20
x_bar = 9800.0      # media muestral (horas)
s = 300.0           # desviación estándar muestral (horas)
mu0 = 10000.0       # valor bajo H0 (horas)
alpha = 0.05        # nivel de significancia (bilateral)

# --------------------------
# 2) Cálculos (t de Student)
df = n - 1
se = s / np.sqrt(n)                         # error estándar de la media
t_stat = (x_bar - mu0) / se                 # estadístico t observado
t_crit = stats.t.ppf(1 - alpha/2, df=df)    # valor crítico (positivo) para prueba bilateral
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df=df))  # p-valor bilateral

# --------------------------
# 3) Impresión paso a paso y explicación
print("PUNTO 30 - PRUEBA t BILATERAL (σ desconocida)")
print("------------------------------------------------")
print(f"H0: μ = {mu0:.1f} horas")
print(f"H1: μ ≠ {mu0:.1f} horas  (prueba bilateral)")
print(f"Datos: n = {n}, x̄ = {x_bar:.1f} h, s = {s:.1f} h, α = {alpha}")
print(f"Grados de libertad: df = n - 1 = {df}")
print(f"Error estándar (SE) = s / √n = {se:.4f} horas")
print(f"Estadístico t observado = (x̄ - μ0) / SE = {t_stat:.4f}")
print(f"Valor crítico t (bilateral, ±) = ±{t_crit:.4f}")
print(f"p-valor (bilateral) = {p_value:.6f}")

# Decisión con la regla de las regiones críticas
if abs(t_stat) > t_crit:
    decision = "Rechazamos H0"
else:
    decision = "No rechazamos H0"

print(f"\nDecisión (α = {alpha}): {decision}")

# Interpretación práctica
if p_value < alpha:
    print(f"Conclusión: Hay evidencia estadística (α = {alpha}) de que la vida media de los focos difiere de {mu0:.0f} h.")
    if t_stat < 0:
        print(f"  Observación práctica: la media muestral = {x_bar:.1f} h < {mu0:.0f} h → los focos duran MENOS de lo declarado.")
    else:
        print(f"  Observación práctica: la media muestral = {x_bar:.1f} h > {mu0:.0f} h → los focos duran MÁS de lo declarado.")
else:
    print(f"Conclusión: No hay evidencia suficiente (α = {alpha}) para afirmar que la vida media difiera de {mu0:.0f} h.")

# --------------------------
# 4) Gráfica: densidad t de x̄ bajo H0 (t con loc=mu0 y scale=SE), regiones de rechazo y t observado
x_min = mu0 - 6*se
x_max = mu0 + 6*se
x = np.linspace(x_min, x_max, 800)
pdf = stats.t.pdf(x, df=df, loc=mu0, scale=se)

plt.figure(figsize=(10,5))
plt.plot(x, pdf, lw=2, label=f'Distribución de x̄ bajo H0: t(df={df}, loc={mu0:.1f}, scale=SE={se:.3f})')

# regiones de rechazo (dos colas)
left_limit = mu0 - t_crit * se
right_limit = mu0 + t_crit * se
plt.fill_between(x, pdf, where=(x <= left_limit), color='red', alpha=0.3, label='Región de rechazo (cola izquierda)')
plt.fill_between(x, pdf, where=(x >= right_limit), color='red', alpha=0.3, label='Región de rechazo (cola derecha)')

# marcar media observada y límites críticos
plt.axvline(x_bar, color='blue', linestyle='--', lw=2, label=f'x̄ observado = {x_bar:.1f} h (t = {t_stat:.3f})')
plt.axvline(mu0, color='black', linestyle='--', label=f'μ0 = {mu0:.1f} h')
plt.axvline(left_limit, color='red', linestyle='--', lw=1.2, label=f'Límite crítico inf = {left_limit:.1f} h')
plt.axvline(right_limit, color='red', linestyle='--', lw=1.2, label=f'Límite crítico sup = {right_limit:.1f} h')

plt.title('Prueba t bilateral para la media - Punto 30')
plt.xlabel('Vida media (horas)')
plt.ylabel('Densidad')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Punto 31 - Prueba χ² para la varianza (prueba bilateral)
# Enunciado: Se afirma σ²0 = 2.5 (minutos^2). Muestra: n = 16, s² = 4. α = 0.05.
# Pregunta: ¿Hay evidencia al 5% de que la varianza poblacional difiere de 2.5?

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# --------------------------
# 1) Datos del problema
sigma2_0 = 2.5     # varianza propuesta bajo H0 (minutos^2)
s2 = 4.0           # varianza muestral observada (minutos^2)
n = 16             # tamaño de la muestra
alpha = 0.05       # nivel de significancia (bilateral)

# --------------------------
# 2) Planteamiento de hipótesis
# H0: σ^2 = sigma2_0
# H1: σ^2 ≠ sigma2_0   (prueba bilateral)

# --------------------------
# 3) Estadístico de prueba (chi-cuadrado)
df = n - 1
chi2_stat = (df * s2) / sigma2_0   # estadístico χ² = (n-1)*s^2 / σ0^2

# Valores críticos para prueba bilateral
chi2_lower = stats.chi2.ppf(alpha/2, df)    # límite inferior
chi2_upper = stats.chi2.ppf(1 - alpha/2, df) # límite superior

# p-valor bilateral: 2 * min{ P(Χ² ≤ chi2_stat), P(Χ² ≥ chi2_stat) }
# equivalente: 2 * min( cdf, 1-cdf )
cdf_at_stat = stats.chi2.cdf(chi2_stat, df)
p_value = 2 * min(cdf_at_stat, 1 - cdf_at_stat)

# --------------------------
# 4) Imprimir resultados paso a paso
print("PUNTO 31 - PRUEBA χ² PARA LA VARIANZA (BILATERAL)")
print("-------------------------------------------------")
print(f"H0: σ² = {sigma2_0:.3f}    vs    H1: σ² ≠ {sigma2_0:.3f}")
print(f"Datos: n = {n}, s² = {s2:.3f}, df = {df}, α = {alpha}")
print()
print(f"Estadístico χ² = (n-1) * s² / σ0² = {df} * {s2:.3f} / {sigma2_0:.3f} = {chi2_stat:.4f}")
print(f"Valores críticos (χ² con df={df}):")
print(f"  Límite inferior  (α/2) : χ²_{{{alpha/2}}} = {chi2_lower:.4f}")
print(f"  Límite superior  (1-α/2): χ²_{{{1-alpha/2}}} = {chi2_upper:.4f}")
print()
print(f"p-valor (bilateral) = {p_value:.6f}")

# Decisión basada en región crítica
if (chi2_stat <= chi2_lower) or (chi2_stat >= chi2_upper):
    decision = "Rechazamos H0"
else:
    decision = "No rechazamos H0"

print(f"\nDecisión (región crítica, α = {alpha}): {decision}")

# Interpretación práctica
if decision.startswith("Rechazamos"):
    print("Conclusión: Hay evidencia estadística al 5% de que la varianza poblacional DIFIERA de 2.5 minutos².")
else:
    print("Conclusión: No hay evidencia suficiente al 5% para afirmar que la varianza poblacional difiera de 2.5 minutos².")

# --------------------------
# 5) Gráfica: densidad χ² y regiones de rechazo, marcando estadístico observado
x_min = max(0, chi2_lower - (chi2_upper - chi2_lower))  # un poco de margen a la izquierda
x_max = chi2_upper + (chi2_upper - chi2_lower)
x = np.linspace(0, x_max, 1000)
pdf = stats.chi2.pdf(x, df)

plt.figure(figsize=(10,5))
plt.plot(x, pdf, lw=2, label=f'Densidad χ² (df={df})')
# sombrear regiones de rechazo (dos colas)
plt.fill_between(x, pdf, where=(x <= chi2_lower), color='red', alpha=0.3, label=f'Región rechazo izquierda (≤ {chi2_lower:.2f})')
plt.fill_between(x, pdf, where=(x >= chi2_upper), color='red', alpha=0.3, label=f'Región rechazo derecha (≥ {chi2_upper:.2f})')
# marcar estadístico observado
plt.axvline(chi2_stat, color='blue', linestyle='--', lw=2, label=f'χ² observado = {chi2_stat:.4f}')
# marcar críticos
plt.axvline(chi2_lower, color='red', linestyle='--', lw=1.2, label=f'Crítico inf = {chi2_lower:.4f}')
plt.axvline(chi2_upper, color='red', linestyle='--', lw=1.2, label=f'Crítico sup = {chi2_upper:.4f}')

plt.title('Prueba χ² para la varianza - Punto 31')
plt.xlabel('χ²')
plt.ylabel('Densidad')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Punto 32 - Prueba χ² para la varianza (unilateral derecha)
# Enunciado: σ0^2 = 1.2 ml^2 (máquina embotelladora).
# Muestra: n = 20, s^2 = 2.0. Nivel de significancia α = 0.01.
# Hipótesis:
#   H0: σ^2 = 1.2
#   H1: σ^2 > 1.2   (queremos probar si la varianza es mayor)

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# --------------------------
# 1) Datos del problema
sigma2_0 = 1.2    # varianza bajo H0 (ml^2)
s2 = 2.0          # varianza muestral observada (ml^2)
n = 20            # tamaño de la muestra
alpha = 0.01      # nivel de significancia (unilateral derecha)

# --------------------------
# 2) Estadístico de prueba χ²
df = n - 1
chi2_stat = (df * s2) / sigma2_0   # χ² = (n-1)*s^2 / σ0^2

# --------------------------
# 3) Valor crítico y p-valor (unilateral derecha)
# Región de rechazo: χ² ≥ χ²_{1-α, df}
chi2_crit = stats.chi2.ppf(1 - alpha, df)
p_value = 1 - stats.chi2.cdf(chi2_stat, df)  # cola derecha

# --------------------------
# 4) Impresión detallada paso a paso
print("PUNTO 32 - PRUEBA χ² (UNILATERAL DERECHA)")
print("--------------------------------------------------")
print(f"H0: σ² = {sigma2_0:.3f}    vs    H1: σ² > {sigma2_0:.3f}")
print(f"Datos: n = {n}, s² = {s2:.3f}, df = n-1 = {df}, α = {alpha}")
print()
print(f"Estadístico χ² observado = (n-1)*s² / σ0² = {df} * {s2:.3f} / {sigma2_0:.3f} = {chi2_stat:.4f}")
print(f"Valor crítico (χ²_{{1-α, df}}) = {chi2_crit:.4f}")
print(f"p-valor (cola derecha) = P(Χ² ≥ {chi2_stat:.4f}) = {p_value:.6f}")
print()

# Decisión
if chi2_stat >= chi2_crit:
    decision = "Rechazamos H0"
else:
    decision = "No rechazamos H0"
print(f"Decisión (nivel α = {alpha}): {decision}")
if decision.startswith("Rechazamos"):
    print("Conclusión: Hay evidencia estadística para afirmar que la varianza es mayor que 1.2 (al nivel especificado).")
else:
    print("Conclusión: No hay evidencia suficiente para afirmar que la varianza sea mayor que 1.2 (al nivel especificado).")

# --------------------------
# 5) Gráfica: densidad χ² con región de rechazo a la derecha y marcando χ² observado
x_max = max(chi2_crit + (chi2_crit*0.5), chi2_stat*1.1)
x = np.linspace(0, x_max, 1000)
pdf = stats.chi2.pdf(x, df)

plt.figure(figsize=(10,5))
plt.plot(x, pdf, lw=2, label=f'Densidad χ² (df={df})')
# sombrear cola derecha (región de rechazo)
plt.fill_between(x, pdf, where=(x >= chi2_crit), color='red', alpha=0.3, label=f'Región de rechazo (χ² ≥ {chi2_crit:.3f})')
# marcar χ² observado y crítico
plt.axvline(chi2_stat, color='blue', linestyle='--', lw=2, label=f'χ² observado = {chi2_stat:.4f}')
plt.axvline(chi2_crit, color='red', linestyle='--', lw=1.5, label=f'Crítico (1-α) = {chi2_crit:.4f}')

plt.title('Prueba χ² para varianza (unilateral derecha) - Punto 32')
plt.xlabel('χ²')
plt.ylabel('Densidad')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Punto 33 - Prueba χ² para la varianza (unilateral izquierda)
# Enunciado: Se sostiene que σ² ≥ 0.5 mm² (la varianza es al menos 0.5).
#           Muestra: n = 12, s² = 0.3. ¿Se puede concluir al 5% que la varianza poblacional es MENOR que 0.5?
# Planteamiento natural (para prueba unilateral izquierda):
#   H0: σ² = 0.5    (hipótesis nula puntual, se interpreta como "al menos 0.5" en contexto)
#   H1: σ² < 0.5    (queremos evidencia de que la varianza es menor)
# Procedimiento: usar estadístico χ² = (n-1)*s² / σ0², cola izquierda.

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# --------------------------
# 1) Datos del problema
sigma2_0 = 0.5    # varianza bajo H0 (mm^2)
s2 = 0.3          # varianza muestral observada (mm^2)
n = 12            # tamaño de la muestra
alpha = 0.05      # nivel de significancia (unilateral izquierda)

# --------------------------
# 2) Estadístico χ²
df = n - 1
chi2_stat = (df * s2) / sigma2_0   # χ² = (n-1)*s^2 / σ0^2

# --------------------------
# 3) Valor crítico y p-valor (cola izquierda)
# Región de rechazo (unilateral izquierda): χ² ≤ χ²_{α, df}
chi2_crit_left = stats.chi2.ppf(alpha, df)    # límite inferior (valor crítico)
p_value_left = stats.chi2.cdf(chi2_stat, df)  # P(Χ² ≤ χ²_obs) para la cola izquierda

# --------------------------
# 4) Impresión paso a paso y decisión
print("PUNTO 33 - PRUEBA χ² PARA LA VARIANZA (UNILATERAL IZQUIERDA)")
print("-------------------------------------------------------------")
print(f"H0: σ² = {sigma2_0:.3f}    vs    H1: σ² < {sigma2_0:.3f}")
print(f"Datos: n = {n}, s² = {s2:.3f}, df = {df}, α = {alpha}")
print()
print(f"Estadístico χ² observado = (n-1)*s² / σ0² = {df} * {s2:.3f} / {sigma2_0:.3f} = {chi2_stat:.6f}")
print(f"Valor crítico (cola izquierda) χ²_{{α, df}} = {chi2_crit_left:.6f}")
print(f"p-valor (cola izquierda) = P(Χ² ≤ {chi2_stat:.6f}) = {p_value_left:.6f}")
print()

# Decisión basada en la región crítica y p-valor
if chi2_stat <= chi2_crit_left:
    decision = "Rechazamos H0"
else:
    decision = "No rechazamos H0"

print(f"Decisión (α = {alpha}): {decision}")

# Interpretación práctica
if chi2_stat <= chi2_crit_left:
    print(f"Conclusión: Hay evidencia estadística (α = {alpha}) de que la varianza poblacional es MENOR que {sigma2_0:.3f} mm².")
else:
    print(f"Conclusión: No hay evidencia suficiente (α = {alpha}) para afirmar que la varianza poblacional sea menor que {sigma2_0:.3f} mm².")
print("\nNota: la prueba de varianza es sensible a la normalidad de la variable; confirmar aproximación si es posible.")

# --------------------------
# 5) Gráfica: densidad χ² y sombreado de la cola izquierda con el χ² observado y el crítico
x_max = max(chi2_crit_left * 3, chi2_stat * 2, df*3)  # ajustar rango para buena visualización
x = np.linspace(0, x_max, 1000)
pdf = stats.chi2.pdf(x, df)

plt.figure(figsize=(10,5))
plt.plot(x, pdf, lw=2, label=f'Densidad χ² (df={df})')
# sombrear la cola izquierda (región de rechazo)
plt.fill_between(x, pdf, where=(x <= chi2_crit_left), color='red', alpha=0.3, label=f'Región rechazo (χ² ≤ {chi2_crit_left:.3f})')
# marcar χ² observado y crítico
plt.axvline(chi2_stat, color='blue', linestyle='--', lw=2, label=f'χ² observado = {chi2_stat:.6f}')
plt.axvline(chi2_crit_left, color='red', linestyle='--', lw=1.5, label=f'Crítico izquierda = {chi2_crit_left:.6f}')

plt.title('Prueba χ² para la varianza (unilateral izquierda) - Punto 33')
plt.xlabel('χ²')
plt.ylabel('Densidad')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
